```python
def pip(price, points = 4):
    if points >= len(price):
        print("Number of points cannot be greater than data length")
        return 0
    
    list_points = [0,len(price)-1]
    while len(list_points) < points + 2:
        
        prev_index = 0
        new_point = 0
        for val in list_points[1:]:
            #start from second item
            y2, x2 = price[val],val
            y1, x1 = price[prev_index], prev_index
            prev_index = val

            #get slope data
            fit_data = line((x1,x2),(y1,y2))[1]
            
            #get new max
            max_idx = np.argmax(abs(price[x1:x2+1] - fit_data))
            if max_idx > new_point:
                new_point = max_idx + x1

        #add_to existing points
        list_points = np.append(list_points, new_point)

        #sort
        list_points = np.sort(list_points)

    return list_points

#changes
def _pip(price, points = 4):
    if points >= len(price):
        print("Number of points cannot be greater than data length")
        return 0
    
    list_of_points = {0:price[0],len(price)-1:price[-1]}
    
    while len(list_of_points) < points + 2:
    # add a new point until the list of points is sufficient
        prev_key = 0
        max_idx = 0
        for key in list_of_points.keys():
            if key == 0:
                prev_key = key
            else:   
                x2, y2 = key, list_of_points[key]
                x1, y1 = prev_key, list_of_points[prev_key]
                prev_key = key

                ## fit a line from start to end
                #get slope data
                fit_data = line((x1,x2),(y1,y2))[1]

                #get new max
                _idx = np.argmax(abs(price[x1:x2+1] - fit_data))
                if _idx > max_idx:
                    max_idx = _idx + x1
                    
        list_of_points[max_idx] = price[max_idx]
        
        list_of_points = dict(sorted(list_of_points.items()))
        
    return list_of_points
```

## Using Stationarity to find an optimal number of points 

Using Stationarity to define a good number of PIPs; incresing PIPs and testing for stationarity

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def fitt_line(dict_vals):
    fitted_data = []
    for index, (key, value) in enumerate(dict_vals.items()):
        if index == 0:
            prev_key = key   
        else:
            y2, x2 = value, key
            y1, x1 = dict_vals[prev_key], prev_key
            prev_key = key
            fitted_data.extend(line((x1,x2),(y1,y2))[1][1:])            
    return fitted_data

points = 6
pip_points_data = _pip(price, points = points)

pip_points_x = [date[index] for index in pip_points_data]
pip_points_y = [price[index] for index in pip_points_data]

fitted_line = fitt_line(pip_points_data)
fitted_line = np.insert(fitted_line,0, price[0])
diff = price - fitted_line

t_stat = adfuller(diff)[0]
p_val = adfuller(diff)[1]
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(14, 8))
legend_label = f'P-Value: {p_val:.4f}\nT-Statistic: {t_stat:.4f}'

ax1.scatter(pip_points_x, pip_points_y, linestyle='-', label= f"{points} point PIPs",color = "red", s= 20)
ax1.plot(pip_points_x, pip_points_y, linestyle='-',color = "red", alpha = 0.5)
ax1.plot(date, price, linestyle='-',color = 'blue', alpha = 0.7)

# Create a twin Axes sharing the xaxis for the second asset
ax2.plot(date, diff, linestyle='-', label=legend_label)
ax2.plot(date, np.ones(np.shape(price))*0, linestyle='-', color='blue', alpha =0.3)

ax1.grid(which='major', axis='both', linestyle='--')
ax2.grid(which='major', axis='both', linestyle='--')

# set labels
ax1.set_ylabel("Price")
ax2.set_ylabel(f"Deviation")

ax1.set_xlabel("Time")
ax2.set_xlabel("Time")



ax1.legend(loc=1)
ax2.legend(loc=1)

plt.show();
```

### Testing multiple PIP's to find the best

```python
points = np.arange(1,50)

tstats = []
pvalues= []
for point in points:
    pip_points_data = _pip(price, points = point)

    pip_points_x = [date[index] for index in pip_points_data]
    pip_points_y = [price[index] for index in pip_points_data]

    fitted_line = fitt_line(pip_points_data)
    fitted_line = np.insert(fitted_line,0, price[0])
    diff = price - fitted_line

    t_stat = adfuller(diff)[0]
    p_val = adfuller(diff)[1]
    
    tstats.append(t_stat)
    pvalues.append(p_val)
```

## Using Fractal Dimension to define an Optimal Number of Points

```python
def get_distance(x_values, y_values):
    total_distance = 0
    for i in range(1, len(x_values)):
        distance = np.sqrt((y_values[i] - y_values[i-1])**2 + (x_values[i] - x_values[i-1])**2)
        total_distance += distance
    return total_distance
```

```python
def fractal_dimension(price):
    cache = {0:price[0],len(price):price[-1]}
    distances = {}

    points_list = np.arange(1,len(price)-10)
    for points in points_list:
        
        x_values = pip(price, points)
        y_values = [price[index] for index in x_values]  
 
        distances[points] = get_distance(x_values,y_values)
    
    return distances
```

```python
# Record the start time
start_time = time.time()
z = fractal_dimension(price)
end_time = time.time()
print(start_time-end_time)
```

```python
distances = np.array(list(z.keys()))
points_list = np.flip(np.array(list(z.values())))

x_val = [distances[0],distances[-1]]
y_val = [points_list[0],points_list[-1]]

frac_dim = 1-line(x_val,y_val)[0]
```

```python
fig, (ax) = plt.subplots(figsize=(14, 8))

ax.scatter(distances, points_list, label = "Fractals",  s=30)
ax.plot(x_val,y_val, label = "Fractal Dimension : {:.3f}".format(frac_dim), color = "red")

ax.grid(which='major', axis='both', linestyle='--')

# set labels
ax.set_title("Fractal Dimension")
ax.set_ylabel("Distance")
ax.set_xlabel("Courseness")
ax.legend(loc=1)

plt.show();
```
